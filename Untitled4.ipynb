```python
# Most Impactful U23 Players in La Liga (2024–2025)

## 1. Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

## 2. Create the dataset
data = {
    "Player": ["Lamine Yamal", "Jude Bellingham", "Gavi", "Nico Williams", "Arda Güler"],
    "Age": [16, 21, 20, 21, 19],
    "Goals": [7, 14, 2, 6, 3],
    "Assists": [5, 4, 4, 7, 2],
    "xG": [6.2, 12.8, 1.5, 5.3, 2.1],
    "xA": [4.4, 3.6, 3.9, 6.1, 1.9],
    "Key_Passes": [32, 40, 28, 35, 18],
    "Tackles": [15, 12, 21, 17, 10],
    "Minutes": [1800, 2100, 1600, 1900, 900]
}

df = pd.DataFrame(data)
df["Goal_Involvement_per90"] = ((df["Goals"] + df["Assists"]) / df["Minutes"]) * 90
df.sort_values("Goal_Involvement_per90", ascending=False, inplace=True)
df.reset_index(drop=True, inplace=True)
df
## 3. Visualize Goal Involvement per 90

plt.figure(figsize=(10,6))
sns.barplot(data=df, x="Goal_Involvement_per90", y="Player", palette="coolwarm")
plt.title("Top U23 Players in La Liga (2024–2025) by Goal Involvement per 90", fontsize=14)
plt.xlabel("Goal Involvement per 90 minutes")
plt.ylabel("Player")
plt.tight_layout()
plt.show()
## 4. Optional: Add Ranking

df["Rank"] = df.index + 1
df[["Rank", "Player", "Goal_Involvement_per90"]]

```
